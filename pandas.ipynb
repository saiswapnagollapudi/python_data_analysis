# apply, apply map, map, replace

```python
new_df = pd.DataFrame({'first':['Corey','Jane', 'John'],'email':['coreys@gmail.com','jane@gmail.com','john@gmail.com'],'last':['s','d','g']})
```

```python
#apply is used to call a function on values
new_df['first'].apply(len)
```

```python
def update_email(email):
  return email.upper()
```

```python
new_df['email'].apply(update_email)
```

```python
new_df['email'] = new_df['email'].apply(update_email)
```

```python
new_df
```

```python
new_df['email'] = new_df['email'].apply(lambda x: x.lower())
```

```python
new_df['email'].apply(len)
```

```python
new_df.apply(len) #this just returns number of rows in each column
```

```python
new_df.apply(len, axis = 'columns') #this just returns number of columns in each row
```

```python
new_df.apply(pd.Series.min)  #when we use apply on dataframe it will apply on each series
```

```python
#applymap will apply function to each value in the dataframe
```

```python
new_df.applymap(len)
```

```python
new_df.applymap(str.lower)
```

```python
#map works only on series and used to substitute a value with another value
new_df['first'].map({'Corey':'chris','Jane':'Mary'})
```

```python
new_df['first'].replace({'Corey':'chris','Jane':'Mary'})
```

```python
new_df['first'] = new_df['first'].replace({'Corey':'chris','Jane':'Mary'})
```
